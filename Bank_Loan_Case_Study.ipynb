### **Outliers in this dataset using Z-score Method AND also by using BOX & Scatter Plots**

```python
numeric_columns = new_apd.select_dtypes(include='number')
def detect_outliers_zscore(data, thresh=3):
    z_scores = (data - data.mean()) / data.std()
    return z_scores.abs() > thresh

outliers_counts = numeric_columns.apply(detect_outliers_zscore).sum()
print("Outliers counts for numerical columns:")
print(outliers_counts)
```

```python
import seaborn as sns
sns.boxplot(data=numeric_columns, orient="h")
plt.title("Box Plots for Numerical Columns")
plt.show()
```

### Top 10 Correlations

```python
numeric_columns_only = new_apd.select_dtypes(include='number').drop(columns=['TARGET'])
correlation_with_target = numeric_columns_only.corrwith(new_apd['TARGET']) #correlation with target column

#calculating 10 top correlations
top_correlations = correlation_with_target.abs().nlargest(10)
print("Top 10 correlations with 'TARGET':")
print(correlation_with_target.loc[top_correlations.index])
```

```python
plt.figure(figsize=(10, 8))
sns.heatmap(new_apd[top_correlations.index].corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Top 10 Correlations with TARGET')
plt.show()
```
